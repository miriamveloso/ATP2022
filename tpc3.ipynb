```python
import random

def menu():
    print('''Menu:
(1) Criar lista aleatória
(2) Criar lista
(3) Soma da lista
(4) Média da lista
(5) Maior elemento da lista
(6) Menor elemento da lista
(7) Verificar se a lista está Ordenada por ordem crescente
(8) Verificar se a lista está Ordenada por ordem decrescente
(9) Procurar um elemento na lista
(0) Sair''')
    return

def criarListaRand(tamanho):
    listaRand=[]
    i=0
    while i<tamanho:
        elemRand=random.randint(1,100)
        listaRand.append(elemRand)
        i=i+1
    return listaRand

def criarLista(tamanho):
    listaNew=[]
    i=0
    while i<tamanho:
        elem=int(input("Introduza o "+str(i+1)+" elemento:"))
        listaNew.append(elem)
        i=i+1
    return listaNew

def somaLista(lista):
    soma=0
    for num in lista:
        soma=soma+num
    return soma

def mediaLista(lista):
    media=somaLista(lista)/len(lista)
    return media

def maiorLista(lista):
    maior=lista[0]
    for i in lista:
        if i>maior:
            maior=i
    return maior

def menorLista(lista):
    menor=lista[0]
    for i in lista:
        if i<menor:
            menor=i
    return menor

def crescenteLista(lista):
    valor=True
    i=0
    while i+1<len(lista):
        if lista[i]>lista[i+1]:
            valor=False
        i=i+1
    return valor

def decrescenteLista(lista):
    valor=True
    i=0
    while i+1<len(lista):
        if lista[i]<lista[i+1]:
            valor=False
        i=i+1
    return valor

def pos(lista, elem):
    posLista=[]
    for i in range(len(lista)):
        if lista[i]==elem:
            posLista.append(i+1)
    if len(posLista)==0:
        posLista.append(-1)
    return posLista

menu()
function=int(input("O que pretende fazer?"))
listaRes=[]

while function!=0:
    if function==1:
        listaRes.clear()
        listaRes=criarListaRand(int(input("Qual o tamanho da lista?")))
    if function==2:
        listaRes.clear()
        listaRes=criarLista(int(input("Qual o tamanho da lista?")))
    if function==3:
        print("")
        print("A soma dos elementos da sua lista é:",somaLista(listaRes))
    if function==4:
        print("")
        print("A média dos elementos da sua lista é:",mediaLista(listaRes))
    if function==5:
        print("")
        print("O maior elemento da sua lista é:",maiorLista(listaRes))
    if function==6:
        print("")
        print("O menor elemento da sua lista é:",menorLista(listaRes))
    if function==7:
        if crescenteLista(listaRes)==True:
            print("")
            print("Sim.")
        else:
            print("")
            print("Não.")
    if function==8:
        if decrescenteLista(listaRes)==True:
            print("")
            print("Sim.")
        else:
            print("")
            print("Não.")
    if function==9:
        posLista=pos(listaRes,int(input("Qual o elemento que procura?")))
        if posLista[0]==-1:
            print("")
            print("O elemento que procura não se encontra na lista.")
        else:
            print("")
            print("O elemento que procura encontra-se na(s)",posLista,"posição(ões).")
    function=int(input("O que pretende fazer?"))
print("")
print("A sua lista é:", listaRes)
```
