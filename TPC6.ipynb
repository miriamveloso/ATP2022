```python
import TPC6 as md


print('''Menu:
(1) Número de obras
(2) Lista de obras
(3) Listas ordenadas
(4) Distribuições
(5) Gráficos
(6) Problema da inversão estrutural
(0) Sair''')
function=int(input("O que pretende fazer?"))

while function != 0:
    if function == 1:
        print("\nO número de obras é: "+str(md.contarObras(md.lerObras("..\datasets\obras.csv"))))
    if function ==2:
        print("\nA lista de obras é:")
        md.imprimeObras(md.lerObras("..\datasets\obras.csv"))
    if function ==3:
        print('''\nSub-menu:
        (1) Lista ordenada por título
        (2) Lista ordenada por ano
        (3) Lista ordenada de compositores
        (0) Sair''')
        subfunction=int(input("Que lista pretende ver?"))
        while subfunction != 0:
            if subfunction ==1:
                print("\nLista ordenada por título:")
                print(md.ordenaTitle(md.lerObras("..\datasets\obras.csv")))
            if subfunction ==2:
                print("\nLista ordenada por ano:")
                print(md.ordenaYear(md.lerObras("..\datasets\obras.csv")))
            if subfunction ==3:
                print("\nLista ordenada de compositores:")
                print(md.ordenaComp(md.lerObras("..\datasets\obras.csv")))
            subfunction=int(input("Que lista pretende ver agora?"))
    if function == 4:
        print('''\nSub-menu:
        (1) Distribuição por período
        (2) Distribuição por ano
        (3) Distribuição por compositor
        (0) Sair''')
        subfunction=int(input("Que distribuição pretende ver?"))
        while subfunction != 0:
            if subfunction ==1:
                print("\nDistribuição por período:")
                print(md.distribPeriodo(md.lerObras("..\datasets\obras.csv")))
            if subfunction ==2:
                print("\nDistribuição por ano:")
                print(md.distribAno(md.lerObras("..\datasets\obras.csv")))
            if subfunction ==3:
                print("\nDistribuição por compositor:")
                print(md.distribComp(md.lerObras("..\datasets\obras.csv")))
            subfunction=int(input("Que distribuição pretende ver agora?"))
    if function == 5:
        subfunction=int(input("Que gráfico pretende ver?"))
        print('''\nSub-menu:
        (1) Gráfico por período
        (2) Gráfico por ano
        (3) Gráfico por compositor
        (0) Sair''')
        while subfunction != 0:
            if subfunction ==1:
                md.graph(md.distribPeriodo(md.lerObras("..\datasets\obras.csv")))
            if subfunction ==2:
                md.graph(md.distribAno(md.lerObras("..\datasets\obras.csv")))
            if subfunction ==3:
                md.graph(md.distribComp(md.lerObras("..\datasets\obras.csv")))
            subfunction=int(input("Que gráfico pretende ver agora?"))
    if function == 6:
        print("\nTabela de obras por autor:")
        md.lerProb(md.prob(md.lerObras("..\datasets\obras.csv")))
    
    function=int(input("O que pretende fazer agora?"))
```
